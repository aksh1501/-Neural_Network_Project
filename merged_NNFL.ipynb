```
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
num_words = 10000

tokenizer = Tokenizer(num_words=num_words, filters='!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n',
                                   lower=True,split=' ')

tokenizer.fit_on_texts(data[3].values)

vocab_size = len(tokenizer.word_index) + 1

X = tokenizer.texts_to_sequences(data[3].values)

word_index = tokenizer.word_index

print('Found %s unique tokens.' % len(word_index))

max_length_of_text = 100
X = pad_sequences(X, maxlen=max_length_of_text)

print(word_index)
```

```
#load embeddings
import codecs
from tqdm import tqdm
print('loading word embeddings...')
embeddings_index = {}
f = codecs.open('/gdrive/My Drive/wiki.simple.vec', encoding='utf-8')
for line in tqdm(f):
    values = line.rstrip().rsplit(' ')
    word = values[0]
    coefs = np.asarray(values[1:], dtype='float32')
    embeddings_index[word] = coefs
f.close()
print('found %s word vectors' % len(embeddings_index))
```

```
from numpy import zeros
embedding_matrix = zeros((vocab_size, 300))
for word, i in tokenizer.word_index.items():
	embedding_vector = embeddings_index.get(word)
	if embedding_vector is not None:
		embedding_matrix[i] = embedding_vector
```

```
import keras
import matplotlib.pyplot as plt
from keras.callbacks import EarlyStopping
class PlotAcc(keras.callbacks.Callback):
    def on_train_begin(self, logs={}):
        self.i = 0
        self.x = []
        self.acc = []
        self.val_acc= []
        self.loss = []
        self.val_loss= []
        
        self.fig = plt.figure()
        
        self.logs = []

    def on_epoch_end(self, epoch, logs={}):
        
        self.logs.append(logs)
        self.x.append(self.i)
        self.acc.append(logs.get('acc'))
        self.val_acc.append(logs.get('val_acc'))
        self.loss.append(logs.get('loss'))
        self.val_loss.append(logs.get('val_loss'))
        self.i += 1
        
        clear_output(wait=True)
        print('val_acc: ', self.val_acc[-1:])
        print('val_loss: ', self.val_loss[-1:])
        
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7))
        ax1.plot(self.x, self.acc, label="acc")
        ax1.plot(self.x, self.val_acc, label="val_acc")
        ax2.plot(self.x, self.loss, label="loss")
        ax2.plot(self.x, self.val_loss, label="val_loss")
        
        ax1.legend()
        ax2.legend()
        plt.show();
        
plot_acc = PlotAcc()
```
